```python
# The merge function to find matches based on common columns
inner = pd.merge(UTC[['UniqueID', 'Case No', 'Receive Time (Badger)', 'Date',
       'Reported Condition', 'Forename', 'Surname', 'DOB', 'Provider Group',
       'Reg Type', 'PG Code', 'Call Origin', 'Initial Case Type',
       'Final Case Type', 'Reception Priority', 'Priority After Assessment',
       'Location Name', 'Booked Date', 'PCC Arrival', 'Despatch Time',
       'Followup', 'Externally Received', 'Destination Name', 'Ext Case ID']], call[['UniqueID', 'Case No', 'Receive Time (Badger)', 'Date',
       'Reported Condition', 'Forename', 'Surname', 'DOB', 'Provider Group',
       'Reg Type', 'PG Code', 'Call Origin', 'Initial Case Type',
       'Final Case Type', 'Reception Priority', 'Priority After Assessment',
       'Location Name', 'Booked Date', 'PCC Arrival', 'Despatch Time',
       'Followup', 'Externally Received', 'Destination Name', 'Ext Case ID']], on=['UniqueID'], how='inner')

print(inner)

```

```python

# Save the merged data to a new CSV file
inner.to_csv(r'C:\Users\Fredrick.Alli\Documents\'AugustInner.csv', index=False)

print("Matched data from Adastra_Report saved to AugustInner.csv")
```

```python
# Read the data into pandas dataframes
call = pd.read_csv(r"C:\Users\Fredrick.Alli\Documents\Cancelled before consultations\Call.csv")
UTC = pd.read_csv(r"C:\Users\Fredrick.Alli\Documents\Cancelled before consultations\UTC.csv")

# To check if unique IDs are present in both dataframes
unique_ids_in_both = call['UniqueID'].isin(UTC['UniqueID']) & UTC['UniqueID'].isin(call['UniqueID'])

# TCreate a new column in call dataframe indicating if unique ID is present in both dataframes
call['Present_in_UTC'] = call['UniqueID'].isin(UTC['UniqueID'])

# To Create a new column in UTC dataframe indicating if unique ID is present in both dataframes
UTC['Present_in_call'] = UTC['UniqueID'].isin(call['UniqueID'])

# Print "Yes" if unique ID is present in both dataframes, otherwise print "No"
for index in call.index:
    print(f"Unique ID {call.loc[index, 'UniqueID']} is present in both dataframes: {'Yes' if unique_ids_in_both[index] else 'No'}")
    
# Save the modified dataframes to CSV files separately
file_path_call = r'C:\Users\Fredrick.Alli\Documents\Cancelled before consultations\Call_modified.csv'
file_path_utc = r'C:\Users\Fredrick.Alli\Documents\Cancelled before consultations\UTC_modified.csv'

call.to_csv(file_path_call, index=False)
UTC.to_csv(file_path_utc, index=False)
```
