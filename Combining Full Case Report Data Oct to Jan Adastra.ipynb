```python
#Importing all the libraries 

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Calling all the CSV files 

oct_23 = pd.read_csv(r"C:\Users\Fredrick.Alli\Documents\2023\10. Oct-23 Full Case Report.csv")
nov_23 = pd.read_csv(r"C:\Users\Fredrick.Alli\Documents\2023\11. Nov-23 Full Case Report.csv")
dec_23 = pd.read_csv(r"C:\Users\Fredrick.Alli\Documents\2023\12. Dec-23 Full Case Report.csv")
jan_24 = pd.read_csv(r"C:\Users\Fredrick.Alli\Documents\2024\1. Jan-24 Full Case Report.csv")

#Shapes
print(oct_23.shape)
print(nov_23.shape)
print(dec_23.shape)
print(jan_24.shape)

#Visaulisation of the data 
oct_23.head()
nov_23.head()
dec_23.head()
jan_24.head()
```

```python
#Printing the number of rows in the table 
#Shapes
print("The number of rows for table A is:", oct_23.shape)
print("The number of rows for table B is:", nov_23.shape)
```

Using Union all/Concat Oct_23 and Nov_23

```python
#setting the max rows 
pd.set_option('display.max_columns', None)

#Printing the shapes 
print("A:", oct_23.shape)
print("B:", nov_23.shape)
print("C:", dec_23.shape)
print("D:", jan_24.shape)

#Union all/concat 

# Concatenate the two DataFrames vertically (union all)
Full_Case_Report_Data = pd.concat([oct_23, nov_23, dec_23, jan_24], ignore_index=True)

# Printing the number of rows in the combined DataFrame
print("The number of rows for the combined table is:", Full_Case_Report_Data.shape[0])
```

```python
# To specify the path in which I want to save this data
Downloads = 'Full_Case_Report_Data.csv'

# To save the combined data frame as a CSV file
Full_Case_Report_Data.to_csv(Downloads, index=False)  # Set index=False to exclude the index column

#To print if the data has been saved 
print(f"The combined data has been saved as '{Downloads}'")
```
