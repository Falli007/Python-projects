```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

matches = pd.read_csv(r"C:\Users\Fredrick.Alli\Documents\Cancelled before consultations\Oct_match 2023.csv")

# Define age groups
age_groups = pd.cut(matches['Age'], bins=[0, 20, 30, 40, 50, 60, float('inf')], labels=['0-20', '21-30', '31-40', '41-50', '51-60', '61+'])

# Calculate the percentage of each age group
percentage_values = age_groups.value_counts(normalize=True) * 100

# Sort by age groups
percentage_values = percentage_values.sort_index()

# Plotting the horizontal bar chart
ax = percentage_values.plot(kind='barh')
plt.xlabel('Percentage')
plt.ylabel('Age Group')
plt.title('Age Group Statistics')

# Displaying the plot
plt.tight_layout()  
plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

# Download the 'punkt' tokenizer
nltk.download('punkt')

# Assuming 'Reported Condition' is the column containing the reported conditions
reported_conditions = matches['Received From'].dropna().str.lower()

# Concatenate all the text into a single string
all_text = ' '.join(reported_conditions)

# Tokenize the text into individual words
words = nltk.word_tokenize(all_text)

# Remove stopwords (common words that do not contribute to the meaning)
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]

# Count the frequency of each word
word_counts = Counter(filtered_words)

# Print the most common words and their frequencies
most_common_words = word_counts.most_common(30)  # Change 10 to the desired number of most common words
for word, count in most_common_words:
    print(f"{word}: {count}")

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)

# Plot the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
